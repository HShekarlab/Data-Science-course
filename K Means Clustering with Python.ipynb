```python
from sklearn.cluster import KMeans
```

```python
k_means_4_data_1 = KMeans(n_clusters = 4)
```

```python
k_means_4_data_1.fit(data_1[0])
```

```python
k_means_4_data_1.cluster_centers_
```

```python
k_means_4_data_1.labels_
```

```python
fig, (axes_1, axes_2) = plt.subplots(1, 2, figsize=(10, 5))

axes_1.set_title("K Means 4")
axes_1.scatter(x= data_1[0][:, 0], y= data_1[0][:, 1], c= k_means_4_data_1.labels_, cmap= "tab10")

axes_2.set_title("Original")
axes_2.scatter(x= data_1[0][:, 0], y= data_1[0][:, 1], c= data_1[1], cmap= "tab10")
```

```python
fig, (axes_1, axes_2) = plt.subplots(1, 2, figsize=(10, 5))

plt.subplot(1, 2, 1)
axes_1.set_title("K Means 4")
sns.scatterplot(x= data_1[0][:, 0], y= data_1[0][:, 1], hue= k_means_4_data_1.labels_, palette= "flare")

plt.subplot(1, 2, 2)
axes_2.set_title("Original")
sns.scatterplot(x= data_1[0][:, 0], y= data_1[0][:, 1], hue= data_1[1], palette= "flare")
```
